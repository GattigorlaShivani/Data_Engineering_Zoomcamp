```python
print(pd.io.sql.get_schema(df, name='green_tripdata', con=engine))
```

```python
df_iter = pd.read_csv('green_tripdata_2019-01.csv', iterator=True, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
len(df)
```

```python
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
```

```python
df.head(n=0).to_sql(name='green_tripdata', con=engine, if_exists='replace')
```

```python
%time df.to_sql(name='green_tripdata', con=engine, if_exists='append')
```

```python
from time import time
```

```python
# loadding green data to db
while True: 
    try:
        t_start = time()

        df = next(df_iter)

        df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
        df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
        
        df.to_sql(name='green_tripdata', con=engine, if_exists='append')

        t_end = time()

        print('inserted another chunk, took %.3f second' % (t_end - t_start))
        
    except StopIteration:
        print("Finished ingesting data into the postgres database")
        break
```

```python
query = """
SELECT * FROM green_tripdata LIMIT 10
"""
pd.read_sql(query, con=engine)
```

```python
# Q3 solution: Count records
query= """SELECT count(*)
FROM green_tripdata 
WHERE lpep_pickup_datetime>='2019-01-15 00:00:00' AND lpep_dropoff_datetime<='2019-01-15 23:59:59'
"""
pd.read_sql(query, con=engine)
```

```python
#Q4 solution : Largest trip for each day
query= """SELECT lpep_pickup_datetime, trip_distance
FROM green_tripdata 
ORDER BY trip_distance DESC
"""
pd.read_sql(query, con=engine)
```

```python
## Q5 solution : The number of passengers
query= """
SELECT passenger_count, count(*)
FROM green_tripdata
WHERE lpep_pickup_datetime>='2019-01-01 00:00:00' and lpep_pickup_datetime <= '2019-01-01 23:59:59'
GROUP BY passenger_count
"""
pd.read_sql(query, con=engine)
```

```python
#loading zonal data to db
df_zone=pd.read_csv("taxi+_zone_lookup.csv")
df_zone.to_sql(name='zones', con=engine, if_exists='replace')
```

```python
## Q6 solution:  Largest tip

query= '''SELECT * FROM Zones WHERE "Zone"='Astoria' '''
pd.read_sql(query, con=engine)
```

```python
query= 'SELECT "DOLocationID" FROM green_tripdata WHERE "PULocationID"=7 ORDER BY tip_amount DESC LIMIT 1'

pd.read_sql(query, con=engine)
```

```python
query= 'SELECT "Zone" FROM Zones WHERE "LocationID"=146'
pd.read_sql(query, con=engine)
```

```python
## Q6 solution in single Query

query =''' SELECT "Zone" 
FROM Zones 
WHERE "LocationID" IN (SELECT "DOLocationID" 
                        FROM green_tripdata 
                        WHERE "PULocationID" 
                        IN (SELECT "LocationID" 
                            FROM Zones 
                            WHERE "Zone"='Astoria') 
                        ORDER BY tip_amount DESC LIMIT 1) '''
pd.read_sql(query, con=engine)
```
